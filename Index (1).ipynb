# Welcome to Jupyter!

```python
import pandas
```

```python
dates = pandas.read_csv('http://webstat.banque-france.fr/fr/downloadFile.do?id=5385698&exportType=csv', delimiter= ';')
```

```python
dates = dates.drop([0,1,2,3,4,5])
```

```python
dates.head()
```

```python
dates = dates.melt(id_vars=['Titre :'], var_name= 'cur', value_name= 'one_euro_value')
```

```python
dates.head(10)
```

```python
dates = dates[(dates.one_euro_value != '-') & (dates.one_euro_value.notnull())]
```

```python
dates.head(20)
```

```python
dates = dates.groupby(['cur']).max()
```

```python
dates.head()
```

```python
dates = dates[['cur', 'Titre :']]
```
